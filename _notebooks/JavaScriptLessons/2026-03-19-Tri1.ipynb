<h2>Trimester in Review</h2>

<p>This trimester went far beyond learning syntax or writing JavaScript. The biggest lesson was developing a coder's mindset—breaking problems into manageable steps, failing fast to learn quickly, and tracking down elusive syntax errors in large files. CSSE 1 taught me a set of habits and tools that I'll carry forward long after the class is over.</p>

<h3>Key Takeaways</h3>
<ul>
<li>Worked comfortably with an operating system and developer toolchain.</li>
<li>Balanced multiple roles: teacher, collaborator, and learner.</li>
<li>Embraced patience and incremental progress instead of rushing to finish.</li>
<li>Planned projects more realistically—and learned from the times I didn't.</li>
</ul>
<p>Below are some highlights from the journey.</p>

<p>One of my very first projects this trimester was the <strong>All About Me</strong> page—a personal introduction built right in this portfolio repo. It was my first real hands-on experience with HTML, CSS, and deploying a live webpage through GitHub Pages. Building it taught me the basics of structuring a page, styling content, and pushing changes to see them go live.
