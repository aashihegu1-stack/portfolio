```python
## Class Homework Challenge

Create a simple JavaScript class with **two different methods**. Instantiate the class and invoke both methods to see their output in the console. A starter code runner is provided below.

```

```python
%%js

// CODE_RUNNER: Define a class with two methods, then create an instance and call them.

class MyClass {
    constructor(name) {
        this.name = name;
    }

    greet() {
        console.log(`Hello, I am ${this.name}`);
    }

    farewell() {
        console.log(`Goodbye from ${this.name}`);
    }
}

const instance = new MyClass("Sample");
instance.greet();
instance.farewell();
```
